```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp ('I will create company for freelancing platform')
```

```python
doc.cats
```

```python
doc[0]
```

```python
doc.text
```

```python
doc[1].pos_
```

```python
doc[5].pos_
```

```python
doc[5].tag_
```

```python
spacy.explain('VBG')
```

```python
spacy.explain('AUX')
```

```python
def showPOS(doc):
  for w in doc :
    print(w,' -- > ',w.pos_,'--->',w.tag_,'--- >', spacy.explain(w.tag_))
showPOS(doc)
```

```python
doc1 = nlp('he left the game')
doc2 = nlp('he is to the left of shelf')
# left is used in different sense in both sentances
```

```python
showPOS(doc1)
```

```python
showPOS(doc2)
```

```python
doc3 = nlp ('the quick brown fox jumps over the lazy dog')
```

```python
from spacy import displacy
```

```python
options = {"ents": ["PERSON", "ORG", "PRODUCT"],"bg": 'yellow',"compact" :True,"distance" :120}
displacy.render(doc3,style ='dep',jupyter = True,options = options)
```
