```python
# Fetch data from 1 to 428 pages and merge/concat into single dataset

df = pd.DataFrame()  # create empty DataFrame
for i in range(1, 429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=d3cab894e1a16c274989f57099adc754&language=en-US&page={}'.format(i))
    #make sure to replace (api_key= d3cab894e1a16c274989f57099adc754) with your api key, this one will not work; you can get one from TMDB's website

    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = pd.concat([df, temp_df], ignore_index=True)

print(df.shape)
```
